## SOAL 1
**Analisis Ukuran Pemusatan Data Tinggi Badan Siswa**
Data tinggi badan (dalam cm) dari 15 siswa di sebuah kelas adalah:

160, 165, 170, 175, 180, 165, 170, 175, 180, 160, 170, 175, 165, 160, 180

Lakukan analisis berikut:

1.  Hitung mean (rata-rata) dari data tinggi badan tersebut.
2.  Hitung median (nilai tengah) dari data tinggi badan tersebut.
3.  Tentukan modus (nilai yang paling sering muncul) dari data tinggi badan tersebut.
4.  Interpretasikan apa yang digambarkan oleh ukuran pemusatan data ini tentang distribusi tinggi badan di kelas tersebut.

## JAWABAN SOAL 1


```
#1. mean (rata-rata)
mean = jumlah seluruh data / banyaknya data
     = (160 + 165 + 170 + 175 + 180 + 165 + 170 + 175 + 180 + 160 + 170 + 175 + 165 + 160 + 180) / 15
     = 2550 / 15
     = 170
```


```
#2. median (nilai tengah) #data ganjil
median = 160, 160, 160, 165, 165, 165, 170, 170, 170, 175, 175, 175, 180, 180, 180
median = 170
```


```
#3. modus (nilai yang sering muncul)
data = 160, 165, 170, 175, 180, 165, 170, 175, 180, 160, 170, 175, 165, 160, 180

#kemunculan > 1 kali
160 → 3 kali
165 → 3 kali
170 → 3 kali
175 → 3 kali
180 → 3 kali

jadi modusnya yaitu 160, 165, 170, 175, dan 180.
```


```
#4. interpretasi data
  a. mean = 170, berarti rata-rata tinggi badan siswa yaitu 170 cm.
  b. median = 170, berarti setengah siswa dikelas memiliki tinggi badan <=170 dan setengah lagi memiliki tinggi badan >=170
  c. modus = 160, 165, 170, 175, 180. berarti tidak ada tinggi badan siswayang dominan.

  dari ukuran pemusatan data dapat disimpulkan bahwa tinggi badan siswa tersebar merata atau tidak ada yang dominan karena data yang terkecil dan yang terbesar memiliki tingkat kemunculan yang sama yaitu 3 kali.

```

## SOAL 2
**Menghitung Ukuran Penyebaran Data**

Data berat badan (dalam kg) dari 10 siswa adalah:

50, 55, 60, 65, 70, 55, 60, 65, 70, 75

Lakukan analisis berikut:

1.  Hitung *range* (rentang) dari data berat badan.
2.  Hitung varians (ragam) dari data berat badan.
3.  Hitung standar deviasi (simpangan baku) dari data berat badan.
4.  Jelaskan seberapa besar penyebaran data berat badan ini dari rata-rata (mean).

## JAWABAN SOAL 2


```
#1. range
data  = 50, 55, 60, 65, 70, 55, 60, 65, 70, 75
range = max - min
      = 75 - 50
      = 25
```
```
#2. varians

#langkah 1 (mean)
mean = 50 + 55 + 60 + 65 + 70 + 55 + 60 + 65 + 70 + 75 / 10
     = 625 / 10
     = 62,5

#langkah 2 (selisih)
50 - 62,5 = -12,5
55 - 62,5 = -7,5
60 - 62,5 = -2,5
65 - 62,5 = 2,5
70 - 62,5 = 7,5
55 - 62,5 = -7,5
60 - 62,5 = -2,5
65 - 62,5 = 2,5
70 - 62,5 = 7,5
75 - 62,5 = 12,5

#langkah 3 (kuadrat)
-12,5^2  = 156,25
-7,5^2 = 56,25
-2,5^2 = 6,25
2,5^2  = 6,25
7,5^2  = 56,25
-7,5^2 = 56,25
-2,5^2 = 6,25
2,5^2  = 6,25
7,5^2  = 56,25
12,5^2 = 156,25

Σ (x - μ)² = 562,5

#langkah 4
varians = Σ (x - μ)² / N
        = 562,5 / 10
        = 56,25
```
```
#3. standar deviasi
Standar Deviasi Populasi (σ) = √Varians Populasi = √σ²

σ  = √σ²
   = √56,25
   = 7,5
```
```
#4. penyebaran data
standar deviasi = 7,5 yang berarti berat badan siswa rata-rata menyimpang sebesar 7,5 kg dari mean (62,5 kg). yang berarti penyebaran data tidak terlalu kecil dan terlalu besar karena variasi data cukup merata (sedang) disekitar nilai rata rata.
```

## SOAL 3
**Kombinasi Pemusatan dan Penyebaran Data**

Hasil ujian matematika 12 siswa adalah:

72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110

Lakukan analisis berikut:

1.  Hitung mean (rata-rata), median (nilai tengah), dan modus (nilai yang paling sering muncul).
2.  Hitung range (rentang), variance (ragam), dan standard deviation (simpangan baku).
3.  Buat kesimpulan tentang performa siswa berdasarkan ukuran pemusatan dan penyebaran data.

## JAWABAN SOAL 3
```
1
#1. mean (rata-rata)
mean = jumlah seluruh data / banyaknya data
     = (72 + 75 + 78 + 80 + 85 + 90 + 92 + 95 + 98 + 100 + 105 + 110) / 12
     = 1080 / 12
     = 90

#2. median (nilai tengah) #data genap
median = 72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110
median = 90 + 92 / 2
       = 182 / 2
       = 91

#3. modus (nilai yang sering muncul)
data = 72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110
-> semua data hanya muncul sekali. jadi data tidak memiliki modus.
```
```
2
#1. range
data  = 72, 75, 78, 80, 85, 90, 92, 95, 98, 100, 105, 110
range = max - min
      = 110 - 72
      = 38

#2. varians

#langkah 1 (mean)
mean = 90

#langkah 2 (selisih)
72 - 90 = -18^2 = 324
75 - 90 = -15^2 = 225
78 - 90 = -12^2 = 144
80 - 90 = -10^2 = 100
85 - 90 = -5^2 = 25
90 - 90 = 0^2 = 0
92 - 90 = 2^2 = 4
95 - 90 = 5^2 = 25
98 - 90 = 8^2 = 64
100 - 90 = 10^2 = 100
105 - 90 = 15^2 = 225
110 - 90 = 20^2 = 400

Σ (x - μ)² = 1636

#langkah 3
varians = Σ (x - μ)² / N
        = 1636 / 12
        = 136,33

#3. standar deviasi
Standar Deviasi Populasi (σ) = √Varians Populasi = √σ²

σ  = √σ²
   = √136,33
   = 11,68
```


```
3. kesimpulan
berdasarkan pemusatan data dari mean, dan median menunjukan bahwa rata-rata nilai siswa adalah 90, dan mediannya adalah 91, maka dari itu disimpulkan bahwa nilai semua siswa tergolong tinggi.

berdasarkan penyebaan data, range nilai sebesar 38 dan standar deviasinya 11,68 menunjukan bahwa penyebaran data nilai siswa cukup lebar. artinya meskipun nilai siswa tergolong tinggi namun terdapat perbedaan yang cukup signifikan antar nilai siswa.
```
