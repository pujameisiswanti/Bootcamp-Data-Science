### Soal 3

Buatlah script Python menggunakan library `matplotlib` untuk menghasilkan sebuah bar chart (grafik batang). Data yang akan digunakan adalah sebagai berikut:

*   **Produk:** `['Produk A', 'Produk B', 'Produk C']`
*   **Jumlah Penjualan:** `[150, 200, 120]`

Aturlah grafik tersebut agar memiliki:
1.  Judul: "Penjualan Produk Bulanan"
2.  Label sumbu X: "Nama Produk"
3.  Label sumbu Y: "Jumlah Penjualan"
4.  Warna batang: 'skyblue'

Tampilkan grafik tersebut.

````python
# TULIS JAWABAN KODE ANDA DISINI
# Pastikan program Anda berjalan tanpa error.

import matplotlib.pyplot as plt

fig = plt.figure()

ax = fig.add_subplot()

produk = ['Produk A', 'Produk B', 'Produk C']
jumlah_penjualan = [150, 200, 120]

ax.bar(produk, jumlah_penjualan, color='skyblue')

ax.set_title('Penjualan Produk Bulanan')
ax.set_xlabel('Nama Produk')
ax.set_ylabel('Jumlah Penjualan')

plt.show()
````
