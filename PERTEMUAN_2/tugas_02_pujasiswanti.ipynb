## Soal 2
**Operasi Aritmatika**

Buatlah program Python untuk menghitung total harga belanjaan. Program ini harus:
1. Meminta input jumlah barang yang dibeli (integer) dan harga per barang (float).
2. Menggunakan operator aritmatika untuk menghitung total biaya sebelum dan sesudah diskon.
3. Menampilkan total biaya sebelum diskon, besarnya diskon (jika ada), dan total biaya akhir.

**Output yang diharapkan:**

```
Jumlah barang: 5
Harga per barang: 25000
Total sebelum diskon: 125000
Diskon: 12500
Total setelah diskon: 112500
```

````python
## JAWABAN SOAL 2

jumlah_barang = int(input("Jumlah barang: "))
harga_per_barang = float(input("Harga per barang: "))
total_sebelum_diskon = jumlah_barang * harga_per_barang
diskon = 12500
total_setelah_diskon = total_sebelum_diskon - diskon

print(f"Total sebelum diskon: {total_sebelum_diskon}")
print(f"Diskon: {diskon}")
print(f"Total biaya akhir: {total_setelah_diskon}")
````

## Soal 3
**Operasi Logika dan Perbandingan**

Buatlah program Python yang meminta input dari pengguna berupa tiga nilai: nilai matematika, nilai bahasa Inggris, dan nilai IPA. Program harus:
1. Menggunakan operator logika dan perbandingan untuk menentukan apakah:
   - Semua nilai lebih besar atau sama dengan 75.
   - Rata-rata dari ketiga nilai lebih besar atau sama dengan 75.
2. Cetak nilai dari dua pernyataan logika tersebut sebagai `True` atau `False`.

**Contoh Output (dengan input tertentu):**
```
Masukkan nilai matematika: 80
Masukkan nilai bahasa Inggris: 70
Masukkan nilai IPA: 60
Semua nilai >= 75: False
Rata-rata nilai >= 75: False
```

````python
## JAWABAN SOAL 3

nilai_matematika = int(input("Masukkan nilai matematika: "))
nilai_bahasa_inggris = int(input("Masukkan nilai bahasa inggris: "))
nilai_ipa = int(input("Masukkan nilai IPA: "))
semua_nilai_lebih_besar_sama_dengan_75 = (nilai_matematika >= 75) and (nilai_bahasa_inggris >= 75) and (nilai_ipa >= 75)
rata_rata_nilai_lebih_besar_sama_dengan_75 = ((nilai_matematika + nilai_bahasa_inggris + nilai_ipa) / 3 ) >= 75

print(f"Semua nilai >= 75: {semua_nilai_lebih_besar_sama_dengan_75}")
print(f"Rata-rata nilai >= 75: {rata_rata_nilai_lebih_besar_sama_dengan_75}")
````
