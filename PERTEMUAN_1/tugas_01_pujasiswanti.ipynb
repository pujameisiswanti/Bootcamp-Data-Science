## SOAL

**Analisis Kasus Data Science**

**Kasus:**

Sebuah perusahaan *e-commerce* bernama TokoOnline123 mengalami penurunan jumlah pelanggan yang melakukan pembelian ulang (*repeat purchase*). Berdasarkan laporan awal, angka pembelian ulang menurun sebesar 20% dalam tiga bulan terakhir, meskipun jumlah pengunjung situs web tetap stabil. Pihak manajemen ingin memahami penyebab dari masalah ini dan meminta tim *data science* untuk menganalisis serta memberikan solusi berbasis data.

5.  **Rekomendasi Solusi:** Berdasarkan langkah-langkah analisis yang Anda buat, berikan rekomendasi solusi awal yang dapat diimplementasikan oleh perusahaan.

## JAWABAN SOAL

1. Masalah utama yang dihadapi Perusahaan TokoOnline123 yaitu penurunan angka pembelian ulang (repeat purchase).

2. Data yang dibutuhkan yaitu: data transaksi dari DB, dan data login pelanggan. Alasannya karena dari data transaksi pelanggan dapat dilihat/ recheck apakah benar terdapat penurunan angka pembelian ulang (repeat purchase)?, sedangkan untuk data login pelanggan dapat dilihat/recheck juga apakah pengunjung situs web benar-benar tetap stabil apa tidak?. Data review juga perlu untuk menggambarkan apa saja feedback dari pelanggan;

3. Yang mungkin terjadi adalah beberapa pelanggan melakukan pembelian ulang produk yang bisa untuk jangka panjang atau pelanggan membeli langsung dalam jumlah yang banyak sehingga butuh waktu yang cukup lama untuk repurchase. Penyebab lainnya bisa jadi karena terdapat toko lain yang menjual produk dengan kualitas bahkan dengan harga yang murah juga, maka dari itu jumlah pengunjung situs tetap stabil namun repurchase mengalami penurunan karena pelanggan bisa jadi membandingkannya dengan toko lain. Kemudian trend juga bisa menyebabkan penurunan repurchase contohnya seperti baju, pelanggan sangat suka dengan kualitas serta model bajunya namun seiring berkembangnya trend baju tersebut sudah ketinggalan zaman yang membuat pelanggan sudah tidak ingin membelinya lagi.

4. langkah langkah:
- pengumpulan data: mengumpulkan data transaksi pelanggan, dan data login pelanggan, serta data review pelanggan
- preprocessing data: melakukan preprocessing data agar data siap untuk dianalsis (seperti menangani missing value, normalisasi, penghapusan dupkikat dll).
- Ekplorasi data: menganalisis pola repurchase dari priode ke priode, analisis dai hasil visualisasi data seperti penjualan dari periode sekian ke priode sekian, visualisasi hubungan antara diskon dan transaksi bisa menggunakan matriks korelasi untuk melihat korelasi antara fitur fitur yang ada. dari matriks korelasi tersebut dapat dilihat mana fitur yang berkorelasi kuat dan mana yang lemah, analisis pelanggang baru vs pelanggan lama yang masih aktif, visulisasi tingkat churn atau pelanggan yang sudah tidak aktif lagi, dan visualisai review pelanggan disetiap produk juga sangat peenting untuk melihat apa feedback dari pelanggan apakah pelanggan puas atau kecewa, dari situ kita bisa melihat apa yang menjadi masalah atau penyebab dari penurunan repurchase atau bisa juga mendapatan solusi dari permasalahan yang lain dari menganalisis data data tersebut.

5. rekomendasi solusi:
- untuk pelanggan yang sering repurchase bisa diberikan diskon berupa gratis ongkir, diskon potongan atau bisa juga menyediakan undian menarik.
- untuk pelanggan yang sering mengunjungi situs namun hanya sekali order tanpa repurchase, bisa juga diberikan promo dengan waktu terbatas untuk menarik perhatian pelanggan tersebut.
- melayani pelaggan dengan fastrespon dan ramah (untuk admiin ecommerce)
- menindaklanjuti keluhan pelanggan seperti eview jelek terhadap produk berarti harus memikirkan bagaimana kita bisa memperbaiki kuliatas produk agar pelanggan merasa puas dan merasa nyaman membeli di tokoonline123 dan berujung repurchase.
