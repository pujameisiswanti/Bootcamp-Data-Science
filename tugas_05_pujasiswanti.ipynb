### Soal 2 (Tipe: Kode)

Tulis program Python yang memenuhi kriteria berikut:
1.  Impor modul `math`.
2.  Definisikan variabel global bernama `global_radius` dan beri nilai `7.0`.
3.  Definisikan sebuah fungsi bernama `hitung_luas_lingkaran` yang menerima satu argumen `radius`.    
 *   Di dalam fungsi, definisikan variabel lokal bernama `pesan_proses`  dengan nilai "Menghitung luas lingkaran...".    
 *   Cetak nilai variabel `pesan_proses`.    
 *   Hitung luas lingkaran menggunakan rumus `math.pi * radius**2`.    
 *   Kembalikan (return) nilai luas yang dihitung.
4.  Panggil fungsi `hitung_luas_lingkaran` menggunakan variabel global `global_radius` sebagai argumen.
5.  Cetak hasil yang dikembalikan oleh fungsi dengan format `"Luas lingkaran dengan radius X adalah Y"`, di mana X adalah nilai `global_radius` dan Y adalah hasil hitungan.Pastikan variabel global dan lokal digunakan sesuai definisi.

````python
# TULIS JAWABAN KODE ANDA DISINI (Python)
# Pastikan program Anda berjalan tanpa error.

import math

global_radius = 7.0

def hitung_luas_lingkaran(radius):
  pesan_proses = "Menghitung luas lingkaran..."
  print(pesan_proses)

  luas_lingkaran = math.pi * radius**2
  return luas_lingkaran

luas_lingkaran = hitung_luas_lingkaran(global_radius)
print(f"Luas lingkaran dengan radius {global_radius} adalah {luas_lingkaran}")
````

### Soal 3 (Tipe: Kode)

Perhatikan kode Python berikut:
```python
global_var = "Saya global"
def ubah_dan_cetak():  
  global_var = "Saya lokal di fungsi"  
  print(f"Di dalam fungsi: {global_var}")
  print(f"Sebelum dipanggil: {global_var}")

ubah_dan_cetak()
print(f"Setelah dipanggil: {global_var}")
```

Jelaskan konsep *scope* variabel yang ditunjukkan oleh kode di atas dan prediksikan output yang akan dihasilkan saat kode tersebut dijalankan.

**1. Penjelasan konsep scope variabel dari kode**

Penjelasan dari kode diatas yaitu terdapat variabel global_var yang merupakan variabel global (tidak didalam fungsi) dan terdapat juga variabel global_var didalam fungsi ubah_dan_cetak. jika memanggil fungsi ubah_dan_cetak, oututnya akan sebagai berikut:

```
#output
Di dalam fungsi: Saya lokal di fungsi
Sebelum dipanggil: Saya lokal di fungsi
```

Namun jika kode diatas di running semua makan hasilnya sebagai berikut:


```
#output
Di dalam fungsi: Saya lokal di fungsi
Sebelum dipanggil: Saya lokal di fungsi
Setelah dipanggil: Saya global
```

tercetaknya setelah dipanggil: saya global terjadi karena variabel global_var dipanggil di luar fungsi, maka dari itu variabel global (gloval_var) yang tercetak.


````python
# TULIS JAWABAN KODE ANDA DISINI (Python)
# Pastikan program Anda berjalan tanpa error.

global_var = "Saya global"

#diluar fungsi
print(f"Sebelum dipanggil: {global_var}")

def ubah_dan_cetak():
  global_var = "Saya lokal di fungsi"
  print(f"Di dalam fungsi: {global_var}")

ubah_dan_cetak() #memanggil yang di dalam fungsi

#diluar fungsi
print(f"Setelah dipanggil: {global_var}")
````

```text
Sebelum dipanggil: Saya global
Di dalam fungsi: Saya lokal di fungsi
Setelah dipanggil: Saya global
```
